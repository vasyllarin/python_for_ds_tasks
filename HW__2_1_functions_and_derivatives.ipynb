1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)
print(f"Похідна функції 10x-5: {derivative_f_lin}")
```

```python
x = np.linspace(0, 10)
f = 10*x + 1
df = 10

plt.figure(figsize=(10,5))
plt.plot(x, f, label='f(x) = 10x - 5', color='green')
plt.axhline(df, label='f(x) = 10', color='orange')
plt.title("Графік функції оцінки залежно від годин, витрачених на навчання, та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Оцінка прямо пропорційно залежить від кількості годин, витрачених на навчання - чим більше витрачаєш годин, ти вища оцінка. Кожна наступна година, витрачена на навчання, збільшує оцінку на 10 балів.  

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin = sp.sin(x)
f_ln = sp.ln(x)
derivative_f_sin = sp.diff(f_sin, x)
derivative_f_ln = sp.diff(f_ln, x)
print(f"Похідна функції sin(x): {derivative_f_sin}")
print(f"Похідна функції ln(x): {derivative_f_ln}")
```

```python
x = np.linspace(-10, 10)
plt.figure(figsize=(10,5))
f_sin = np.sin(x)
derivative_f_sin = np.cos(x)
plt.plot(x, f_sin, label='f(x) = sin(x)', color='green')
plt.plot(x, derivative_f_sin, label='f(x) = cos(x)', color='blue')
plt.title("Графік функції sin(x) та її похідної (cos(x))")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(0.1, 10)
plt.figure(figsize=(10,5))
f_ln = np.log(x)
derivative_f_ln = 1 / x
plt.plot(x, f_ln, label='f(x) = ln(x)', color='green')
plt.plot(x, derivative_f_ln, label='f(x) = 1/x', color='blue')
plt.title("Графік функції ln(x) та її похідної (1/x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f_lin = x**2 - 5 * x + 1
derivative_f_lin = sp.diff(f_lin, x)
print(f"Похідна функції x^2 - 5x + 1: {derivative_f_lin}")
```

```python
x = np.linspace(0, 100)
f = x**2 - 5 * x + 1
df = 2*x - 5

plt.figure(figsize=(10,5))
plt.plot(x, f, label='f(x) = x^2 - 5x + 1', color='green')
plt.plot(x, df, label='f(x) = 2x - 5', color='orange')
plt.title("Графік функції витрат залежно від к-ті вироблених одиниць продукції, та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
